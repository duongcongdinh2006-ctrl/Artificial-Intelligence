```python
#14
#b
import numpy as np
m = int(input("Nhap vao so hang: "))
n = int(input("Nhap vao so cot: "))
arr1 = np.random.randn(m,n)
k = int(input(f"Nhap vao dong k (Tu 0 den {m-1}): "))
if 0 <= k < m:
    dong_k = arr1[k, :]
    print(f"Dong thu {k} la: \n{dong_k}")
else:
    print(f"Loi: Chi so dong phai nam trong khoang tu 0 den {m-1}!")
```
